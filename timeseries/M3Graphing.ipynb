```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_tsdata = pd.read_csv('ConstructionTimeSeriesDataV2.csv')
df_tsdata.head

Date = df_tsdata['Month-Year']
TotConstruction = df_tsdata['Total Construction']
TotPrivate = df_tsdata['Private Construction']
TotPublic = df_tsdata['Public Construction']

fig,ax = plt.subplots()

plt.plot(Date, TotPrivate, color = 'b', alpha = 0.75)
plt.plot(Date, TotPublic, color = 'k', alpha = 0.75)

fig.set_size_inches(10,8)
fig.suptitle("Public and Private Construction Spending", fontsize = 20)
ax.xaxis.set_label_text('Date', fontsize = 16)
ax.yaxis.set_label_text('Construction Spending ($)', fontsize = 16)
ax.set_xticks(Date[::6])
ax.set_xticklabels(Date[::6])
#ax.axvline(x = 'Dec-07', linewidth = 1, color = 'red', label = 'Start', linestyle = '--')
#ax.axvline(x = 'Jun-09', linewidth = 1, color = 'red', label = 'Start', linestyle = '--')
ax.axvspan('Dec-07', 'Jun-09', ymin=0, ymax=100000, facecolor = 'k', alpha = 0.1)
plt.xticks(rotation = 'vertical')
ax.legend(loc=1)
ax.text(80, 85000, 'Great\nRecession',fontsize = 12,verticalalignment = 'center', horizontalalignment = 'center')
```

As the warning signs for the sub-prime mortgage crisis/housing bubble bursting started flashing in 2006 thru early 2007, private construction spending plateaued and started then fall.  Once the housing bubble burst and the Great Recession set in, private construction spending dropped precipitously until well after the recession formally ended.  

At the same time, an uptick started happening in public construction.  A lot of this uptick could be associated with the trillion-dollar stimulus package Congress approved to try to staunch the bleeding of the recession.
